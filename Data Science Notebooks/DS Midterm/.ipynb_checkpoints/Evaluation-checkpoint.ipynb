Our code will look at the value/weight ratio to determine which items to add to the knapsack.

It will select the highest ratio items and add to knapsack until the capacity is 100.

```python
import numpy as np

'''Generates 300 capacities for each instance between 100-160. 
Normal distribution with mean: 130 and standard deviation: 30'''

def generate_knap():
    
    f = open("Knap_Sizes.csv", "w") # create file.

    results = np.random.normal(130, 30, 300) # generates capacities.

    instances = []
    for i in results:
        instances.append(int(i))

    for instance in instances:
        f.write(f'{instance},\n')

    f.close()
```

```python
def run_evalution(knap_sizes_file_destination):

    # Read in given instance file with capacities and values of items.
    # Covert to list of lists.

    knapsack_file = open('./Knapsack-Instances.csv')
    lines = knapsack_file.readlines()
    instance_list = [] # [[weights],[values],[blanks],... * 300].
    for line in lines:
        line = line.rstrip()
        instance_list.append(line.split(','))


    #Constructs a master list to select items to add to knapsack.

    master_list = []
    for i in range(0, 900, 3): # Skips to all weight rows.
        test_list = []
        for index, w in enumerate(instance_list[i]): # irirating over individual weight rows.
            weight = int(w)
            value = float(instance_list[i+1][index])
            ratio = value/weight # Measure to be used to select items.
            test_list.append([ratio, weight, value, index]) # Record index to assign items later.
            test_list = sorted(test_list, reverse=True)
        master_list.append([i, test_list]) # 0 - 299


    # Load actual knap sizes to list.

    knap_sizes_file = open(knap_sizes_file_destination)
    knap_lines = knap_sizes_file.readlines()
    knap_sizes = []
    for line in knap_lines:
        line = line.split(',')
        size = int(line[0])
        knap_sizes.append(size)

    knap_sizes_file.close()

    # Select items to add to knapsack based on weight/value ratio
    # Records total capacity used and value stored

    master_solution_list = []
    total_used_C = 0 # total amount of capacity used.
    total_C = 0 # total amount of capacity available.
    total_v = 0 # total value stored safely.
    lost_v = 0 # total value lost to exceeding knap capacity.
    perfect_v = 0 # total value stored safely if instance capacity is known.
    perfect_C = 0 # total capacity used if instance capacity is known.
    for x in range(0, 300):
        test_list = master_list[x][1]
        test_weight = 100
        perfect_weight = knap_sizes[x]
        solution_list = []
        total_C += test_weight
        used_C = 0
        potential_v = 0
        for i, v in enumerate(test_list):
            if test_list[i][1] <= test_weight:
                used_C += v[1]
                potential_v += v[2]
                solution_list.append(v)
                test_weight -= test_list[i][1]
            if test_list[i][1] <= perfect_weight:
                perfect_C += v[1]
                perfect_v += v[2]
                perfect_weight -= test_list[i][1]
        if used_C <= knap_sizes[x]:
            total_v += potential_v
            total_used_C += used_C
        else:
            lost_v += potential_v

        master_solution_list.append([x, solution_list])

    # Generate the solution file and write selected items to it.

    solution_list = []
    for i in range(0, 900, 3):
        solution_list.append(instance_list[i])
        solution_list.append(instance_list[i+1])
        index_list = []
        temp_list = []
        master_index = int(i/3)
        for x in master_solution_list[int(i/3)][1]:
            index_list.append(x[3])
        for x in range(50):
            if x in index_list:
                temp_list.append(str(1))
            else:
                temp_list.append(str(0))
        solution_list.append(temp_list)

    solution_file = open('./knapsack_solution.csv', 'w')

    for line in solution_list:
        solution_string = ''
        for value in line:
            solution_string += f'{value},'
        solution_file.write(f'{solution_string}\n')

    solution_file.close()

    # Print the statistics of heuristic approach

    #print(f'''Total Capacity Used: {total_used_C} vs Perfect: {perfect_C}
    #Total Value Captured: {total_v} vs Perfect: {perfect_v}
    #Total Value Loss to Exceeding Capacity: {lost_v}''')
    
    return total_used_C, perfect_C, total_v, perfect_v

generate_knap()
run_evalution('./Knap_Sizes.csv')
```

```python
# Monte Carlo Simulation run N times

N = 100
heuristic_capacity = []
perfect_capacity = []
heuristic_value = []
perfect_value = []

for trial in range(N):
    generate_knap()
    total_used_C, perfect_C, total_v, perfect_v = run_evalution('./Knap_Sizes.csv')
    heuristic_capacity.append(total_used_C)
    perfect_capacity.append(perfect_C)
    heuristic_value.append(total_v)
    perfect_value.append(perfect_v)

heuristic_capacity = round(np.mean(heuristic_capacity), 2)
perfect_capacity = round(np.mean(perfect_capacity), 2)
heuristic_value = round(np.mean(heuristic_value), 2)
perfect_value = round(np.mean(perfect_value), 2)

print(f'''Over {N} Trials our heuristic used {heuristic_capacity} capacity versus the perfect {perfect_capacity} capacity
and stored {heuristic_value} value safely versus the perfect {perfect_value} value.''')
```
