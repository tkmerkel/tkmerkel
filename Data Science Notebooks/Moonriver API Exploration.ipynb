```python
import requests
import pandas as pd
from datetime import datetime
```

```python
# My address
#address = '0x707fC4CE82Fe8d452715f07338fE77a912807252'

# Random guy's address
address = '0x525f1221728afbc867023cae17a8e4ea0f5aab85'
#address = '0xc0bc7bbc9afa9b0577d10bd3eb636c8d3ab841d3'
#address = '0xf362659f3d2bffd4789d0616ae44c80a108533a1'


api_token = 'KUM5JZVJ8FF46UMJB4KUAB1MQUSYWYI2EJ'

# Single address balance call
# url = f'https://api-moonriver.moonscan.io/api?module=account&action=balance&address={address}&tag=latest&apikey={api_token}'

# Get a list of Normal Transactions By Address
# url = f'https://api-moonriver.moonscan.io/api?module=account&action=txlist&address={address}&startblock=1&endblock=99999999&sort=asc&apikey={api_token}'

# Get Token transfers for an address
url = f'https://api-moonriver.moonscan.io/api?module=account&action=tokentx&address={address}&startblock=0&endblock=2500000&sort=asc&apikey={api_token}'

# Get a list of 'Internal' Transactions by Address
# url = f'https://api-moonriver.moonscan.io/api?module=account&action=txlistinternal&address={address}&startblock=0&endblock=2702578&sort=asc&apikey={api_token}'

address = address.lower()
response = requests.get(url)
data = response.json()

def id_sent_transaction(row):
    if row['to'] == address:
        return 1
    else:
        return -1

df = pd.DataFrame(data['result'])
df['value'] = df['value'].astype(float)
df['tokenDecimal'] = df['tokenDecimal'].astype(float)
df['quantity'] = df['value'] / (10**df['tokenDecimal'])
df['sentDummy'] = df.apply(id_sent_transaction, axis=1)
df['balance'] = df['quantity'] * df['sentDummy']
print(f'Balance of {address}:')
pd.DataFrame(df.groupby('tokenName')['balance'].sum())['balance'].round()
```

```python
df.to_csv('transactions.csv')
```
